## <u>Correlation with Heatmap

    A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discrete dimensions or event types. Its very important in Feature selection. It will only work on numerical values.

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.heatmap(df.corr(),annot=True,cmap='Reds')
```

## <u>Jointplots
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D density plot.

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
## There are various kinds are here,Try them
##    kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
## It helps us to find the higher concentration
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='kde')
```

## Pair plot 
A 'pairs plot' is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs Plots are just eldaborations on this, showing all variables paired with all the other vairables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
## Most of the peoples are non-smokers
```

## <u>Distplot
Distplot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=True,bins=20)
```

## <u>Categorical Plots:
Seaborn also helps us in doing the analysis on categorical Data points. Included plots are,
 - boxplot
 - violinplot
 - countplot
 - barplot

```python
df.dtypes
```

## Count plot

```python
sns.countplot('sex',data=df)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.countplot('sex',data=df)
## Here its shows most of the Males gave tips
```

```python
sns.countplot(x='day',data=df)
## Many people go eat on saturday and sundays
```

```python
sns.countplot(y='sex',data=df)
```

## Bar Plot

```python
sns.barplot(x='total_bill',data=df)
## You have to give two columns for barplot
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

- when males come bill will be around 20
- when females come bill will be around 18

```python
sns.barplot(x='total_bill',y='smoker',hue='sex',data=df)
```

## Box Plot
A box and whisker plot(sometimes called boxplot) is graph that presents information from five-number summery

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(x='total_bill',data=df,y='day',hue='smoker')
```


See Also
--------
- violinplot : A combination of boxplot and kernel density estimation.    
- stripplot : A scatterplot where one variable is categorical. Can be used
            in conjunction with other plots to show each observation.    
- swarmplot : A categorical scatterplot where the points do not overlap. Can
            be used with other plots to show each observation.    
- catplot : Combine a categorical plot with a :class:`FacetGrid`.
