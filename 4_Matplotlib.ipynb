# Matplotlib
 - 2 and 3d visualization tool
 

```python
import matplotlib.pyplot as plt
%matplotlib inline  #in here you dont want to use 'plt.show()' command every time
```

```python
import numpy as np
```

```python
x=np.arange(0,11)
y=np.arange(10,21)
```

# Scatter plot


```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('scatter.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r',linestyle='--',linewidth=4,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title(' 2D Diagram')
```

```python
plt.plot(x,y,'r*--')
```

# Subplots
- you can use multiple diagrams in single plot

```python
## creating subplots

plt.subplot(2,2,1)  ##subplot(nrow,ncol,position)
plt.plot(x,y,'r--')
plt.subplot(2,2,4)
plt.plot(x,y,'g+-')
```

```python
## create complex calculations in ploting variables

x = np.arange(0,4*np.pi,step=0.1)
y=np.sin(x)
plt.title('sine wave')

##plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
## using subplots
x = np.arange(0,5*np.pi,step=0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,2,1)
plt.plot(x,y_sin,'r--')
plt.title('Sine')

plt.subplot(2,2,4)
plt.plot(x,y_cos,'b--')##b-- means color blue and line style is '--'
plt.title('Cosine')


#this is optional because it is inline  matplotlib
plt.show()
```

# Bar plot


```python
x=[2,8,10]
y=[10,12,8]

x1=[3,9,11]
y1=[3,10,5]

plt.bar(x,y,color='g')
plt.bar(x1,y1,color='b')

plt.title('Bar graph')
plt.ylabel('Y label')
plt.xlabel('X label')
```

# Histogram

```python
a=np.array([22,14,2,5,25,36,25,85,47,58,96,41,54,62,57,29,85])
np.sort(a)
plt.hist(a)
#plt.hist(a,bins=5) ##Bins define the no on bins to show
plt.title('Histogram')
#It includes the no.of accurance of value between x to x1 range
```

# Box plot
- syntax
-    plt.boxplot(data,vert(vertical)=True,patch_artist(color)=True)

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

## rectangular Box plot
#plt.boxplot(data,vert=True,patch_artist=True)
plt.boxplot(data,vert=True,patch_artist=False)
##patch_artist=False means it make the box empty without this blue color
plt.show()
```

# Pie chart


```python
#Data's for plot
labels='Python','C++','Ruby','Java'
sizes=[219,120,200,290]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0)  #slicing 1st piece of chart from pie

#plot
plt.pie(sizes)
plt.show()
```

```python
#plot
plt.pie(sizes,explode=explode,shadow=True)
#plt.show() ##If you aren't giving plt.show() It will show some dum values down here.Try to use plt.show() always
```

```python
#include values
plt.pie(sizes,explode=explode,labels=labels,shadow=True,autopct='%1.6f%%') ## change the floating values in this autopct field
plt.axis('equal')
plt.show()
```
